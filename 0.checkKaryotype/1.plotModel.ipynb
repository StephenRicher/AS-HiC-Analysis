```python
import sys
sys.path.append('..')
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from utilities import defaultPlotting, formatCell
```

```python
dtype = {'file': str, 'Diploid': float, 'Triploid': float, 'Tetraploid': float}
df = pd.read_csv('nQuire-allResults.tsv', usecols=[0,5,6,7], skiprows=1, dtype=dtype, names=dtype.keys(),sep='\t')
```

```python
df['cell'] = df['file'].apply(lambda x: formatCell(x.split('/')[-1].split('-')[0]))
df['chrom'] = df['file'].apply(lambda x: x.split('/')[-1].split('-')[1])

allChroms = [f'chr{c}' for c in range(1, 23)] + ['chrX']
```

```python
hb2 = pd.read_csv('../../HB2/nQuire/HB2-nQuire.results.tsv', usecols=[0,5,6,7], skiprows=1, dtype=dtype, names=dtype.keys(),sep='\t')
hb2['cell'] = formatCell('HB2')
hb2['chrom'] = hb2['file'].apply(lambda x: x.split('-')[1])

df = pd.concat([df, hb2])
```

```python
defaultPlotting(width=180, ratio=1)
```

```python
fig, axes = plt.subplots(2, 2)
cbar_ax = fig.add_axes([.91, .3, .03, .4])
axes = axes.flatten()
for i, cell in enumerate(df['cell'].unique()):
    data = df.loc[df['cell'] == cell].drop(['file', 'cell'], axis=1).set_index('chrom').reindex(allChroms)
    data = data.div(data.max(axis=1), axis=0)
    if cell == formatCell('H1hESC'):
        data.loc[data.index == 'chrX'] = np.nan
    cbar = None if i < 3 else cbar_ax
    g = sns.heatmap(
        data, cbar=i==3, cbar_ax=cbar, linewidths=0.1, vmin=0, vmax=1,
        linecolor='black',yticklabels=1, cmap='Reds_r', 
        cbar_kws={'label': 'Delta Log-Likelihood'}, ax=axes[i])
    g.set_xticklabels(g.get_xticklabels(), rotation=0)
    axes[i].set_ylabel('')
    axes[i].set_title(f'{cell}', loc='left')
    axes[i].set_facecolor('gray')
fig.tight_layout(rect=[0, 0, .9, 1])
fig.savefig('nQuire-karyotype-analysis.svg')
```
