Count number of ASE / Imprinted genes overlapping ASTADs and compare to random sample

```python
import numpy as np
import pandas as pd
import seaborn as sns
from pybedtools import BedTool
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from statsmodels.stats.multitest import multipletests
```

```python
plt.rcParams['font.size'] = '15'
```

```python
def getSymbol(x):
    "Extract gene name from GFF3 attributes string"
    attribute = dict([pair.split('=') for pair in x.split(';')])
    return attribute['gene_name']
```

```python
def assignGroup(x):
    if 'ASTAD' in set(x['blockCount']):
        return 'ASTAD'
    else:
        return 'TAD'
```

```python
columns = ['seqname', 'start', 'end', 'name', 'score', 'strand']
gencodeGenes = BedTool('../gencode.v38lift37-genes.gff3').to_dataframe()
gencodeGenes['name'] = gencodeGenes['attributes'].apply(getSymbol)
gencodeGenes = BedTool.from_dataframe(gencodeGenes[columns])
```

```python
allGenes = []
for cell in ['GM12878', 'IMR90', 'H1hESC']:
    TADs = f'../diffTAD/results/{cell}/alleleGRCh37/{cell}_a1-vs-{cell}_a2-all-20000-SNPsplit_diff_tad-allTAD-autosome.bed'
    # Only consider genes overlapping TADs
    geneInTAD = gencodeGenes.intersect(TADs, wa=True, wb=True).to_dataframe()
    # If a gene intersects an ASTAD then assign to ASTAD
    geneInTAD = (
        geneInTAD.groupby(['chrom', 'start', 'end', 'name', 'strand'])
        .apply(assignGroup)
        .reset_index()
        .rename({0: 'group'}, axis=1))
    geneInTAD['cell'] = cell
    geneInTAD['ASTAD'] = geneInTAD['group'] == 'ASTAD'
    aseg = pd.read_csv(f'{cell}-ASEG-GRCh37.bed', sep='\t', header=None)[3]
    geneInTAD['ASEG'] = geneInTAD['name'].isin(aseg)
    allGenes.append(geneInTAD)
allGenes = pd.concat(allGenes)

imprinted = pd.read_csv('Imprinted-GeneImprint-GRCh37.bed', header=None, sep='\t')[3]
allGenes['Imprinted'] = allGenes['name'].isin(imprinted)
```

```python
results = {}
rawP = [] # Store in list for post correction
for cell in ['GM12878', 'IMR90', 'H1hESC']:
    # Consider only genes within a TAD
    genesInTad = allGenes.loc[allGenes['cell'] == cell]
    for geneType in ['ASEG', 'Imprinted']:
        obs = pd.crosstab(genesInTad['ASTAD'], genesInTad[geneType])
        chi2, p, dof, exp = chi2_contingency(obs)
        name = 'ASEG' if geneType != 'Imprinted' else geneType
        results[(cell, name)] =  [chi2, p, np.log2(obs / exp)]
        rawP.append(p)
_, adjP, _, _ = multipletests(rawP, method='bonferroni')
```

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 13.5), sharex=True, sharey=True)
axes = axes.flatten()
for i, (cell, name) in enumerate(results.keys()):
    chi2, p, obsExp = results[(cell, name)]
    p = adjP[i] # Obtain corrected p
    sns.heatmap(
        obsExp, cmap='bwr', center=0, vmin=-1, vmax=1, annot=True, fmt='.3f',
        cbar_kws={'label': 'log2(obs / exp)'}, ax=axes[i])
    if i % 2 != 0:
        axes[i].set_ylabel('')
    if i > 3:
        axes[i].set_xlabel(f'{name}')
    else:
        axes[i].set_xlabel('')
    axes[i].set_title(f'{cell} - {name}, p(adj) = {p:.3f}', loc='left')
fig.tight_layout()
fig.savefig('ASGeneASTADchi2.svg')
```
