## Standardise Gene IDs for Allele Specific Expressed (ASE) and Imprinted gene lists

All gene lists are converted to Ensembl Gene IDs and mapped back Gencode v38 (GRCh37)
Some manual curation was performed where gene symbols are different to those in Gencode v38.
Genes with no appropriate ID, or ambiguous IDs are removed.

```python
import venn
import glob
import subprocess
import numpy as np
import pandas as pd
from pybedtools import BedTool
import matplotlib.pyplot as plt
```

```python
# Read Gencode V38 and extract coordinates with Ensembl ID
gencodeV38path = '../gencode.v38lift37-genes.gff3' # Genes only
gencodeV38 = {'chrom': [], 'start': [], 'end': [], 'ID': [], 'symbol': [], 'orientation': []}
with open(gencodeV38path) as fh:
    for line in fh:
        line = line.split()
        gencodeV38['chrom'].append(str(line[0]))
        gencodeV38['start'].append(int(line[3])) 
        gencodeV38['end'].append(int(line[4]))
        gencodeV38['orientation'].append(line[6])
        attributes = line[8].split(';')
        for a in attributes:
            tag, value = a.split('=')
            if tag == 'gene_name':
                gencodeV38['symbol'].append(value)
            if tag == 'gene_id':
                gencodeV38['ID'].append(value.split('.')[0])
gencodeV38 = pd.DataFrame(gencodeV38)
```

### Remap IMR90 ASE genes

```python
aseIMR90 = 'IMR90-ASEG-GSM438363_GSM432687-GRCh38.xlsx'
aseIMR90 = pd.read_excel(aseIMR90)
```

```python
genV38_IMR90 = pd.merge(gencodeV38, aseIMR90['GeneID'].to_frame(), left_on='symbol', right_on='GeneID')
dupSymbol = genV38_IMR90['symbol'].duplicated(keep=False)
dupID = genV38_IMR90['ID'].duplicated(keep=False)
genV38_IMR90 = genV38_IMR90.loc[~(dupSymbol | dupID)]
genV38_IMR90['score'] = 0
genV38_IMR90['rgb'] = "0,0,0"
nRemoved = len(aseIMR90) - len(genV38_IMR90)
print('Total', len(aseIMR90), 'Removed:', nRemoved, 'Retained:', len(genV38_IMR90))
```

```python
columns = ['chrom', 'start', 'end', 'symbol', 'score', 'orientation', 'start', 'end', 'rgb']
BedTool.from_dataframe(genV38_IMR90[columns]).sort().moveto('IMR90-ASEG-GRCh37.bed')
```

### Remap H1hESC ASE genes

```python
aseH1 = 'H1hESC-ASEG-GSM438361_GSM432685.GRCh38.xlsx'
aseH1 = pd.read_excel(aseH1)
```

```python
genV38_H1 = pd.merge(gencodeV38, aseH1['GeneID'].to_frame(), left_on='symbol', right_on='GeneID', how='right')
dupSymbol = genV38_H1['symbol'].duplicated(keep=False)
dupID = genV38_H1['ID'].duplicated(keep=False)
genV38_H1 = genV38_H1.loc[~(dupSymbol | dupID)]
genV38_H1['score'] = 0
genV38_H1['rgb'] = "0,0,0"
genV38_H1['start'] = genV38_H1['start'].astype(int)
genV38_H1['end'] = genV38_H1['end'].astype(int)
nRemoved = len(aseH1) - len(genV38_H1)
print('Total', len(aseH1), 'Removed:', nRemoved, 'Retained:', len(genV38_H1))
```

```python
columns = ['chrom', 'start', 'end', 'symbol', 'score', 'orientation', 'start', 'end', 'rgb']
BedTool.from_dataframe(genV38_H1[columns]).sort().moveto('H1hESC-ASEG-GRCh37.bed')
```

### Remap GM12878 ASE genes

```python
def defineAllele(x):
    if str(x).lower() == 'paternal':
        return '255,0,0'
    elif str(x).lower() == 'maternal':
        return '0,0,255'
    else:
        return '128,128,128'
```

```python
aseGM12878 = 'GM12878-ASEG-31740818-supplTab11-genes.txt'
names = ['name', 'maternal', 'paternal', 'ASE']
aseGM12878 = pd.read_csv(aseGM12878, names=names, usecols=[0, 2, 3, 5], comment='#', sep='\t')
```

```python
genV38_GM12878 = pd.merge(gencodeV38, aseGM12878[['name', 'ASE']], left_on='ID', right_on='name')
genV38_GM12878['rgb'] = genV38_GM12878['ASE'].apply(defineAllele)
dupSymbol = genV38_GM12878['symbol'].duplicated(keep=False)
dupID = genV38_GM12878['ID'].duplicated(keep=False)
genV38_GM12878 = genV38_GM12878.loc[~(dupSymbol | dupID)]
genV38_GM12878['score'] = 0
nRemoved = len(aseGM12878) - len(genV38_GM12878)
print('Total', len(aseGM12878), 'Removed:', nRemoved, 'Retained:', len(genV38_GM12878))
```

```python
columns = ['chrom', 'start', 'end', 'symbol', 'score', 'orientation', 'start', 'end', 'rgb']
(BedTool.from_dataframe(genV38_GM12878.loc[genV38_GM12878['ASE'] != 'Unassigned', columns])
 .sort().moveto('GM12878-ASEG-GRCh37.bed'))
BedTool.from_dataframe(genV38_GM12878[columns]).sort().moveto('GM12878-all-GRCh37.bed')

# Write custom BED with no gene name for non-ASEG
genV38_GM12878['symbol'] = genV38_GM12878.apply(lambda x: '' if x['rgb'] == '128,128,128' else x['symbol'], axis=1)
genV38_GM12878[columns].to_csv('GM12878-all-noName-GRCh37.bed', header=False, index=False, sep='\t')
```

## Remap Imprinted genes

```python
imprinted = 'GeneImprint.csv'
imprinted = pd.read_csv(imprinted, sep=',', names=['symbol', 'alias', 'location', 'status', 'allele'])
# Remove trailing whitespace
imprinted['symbol'] = imprinted['symbol'].str.strip()
imprinted['status'] = imprinted['status'].str.strip()
```

```python
remap = ({
    'AIM1': 'CRYBG1',
    'SLC22A2*': 'SLC22A2',
    'SLC22A3*': 'SLC22A3',
    'INPP5F V2': 'INPP5F',
    'IGF2AS': 'IGF2-AS',
    'SNORD115@': 'SNORD115-1',
    'PSIMCT-1': 'MCTS2P',
    'MCTS2': 'MCTS2P',
    'GNASAS': 'GNAS-AS1',
})

for name, new in remap.items():
    imprinted.loc[imprinted['symbol'] == name, 'symbol'] = new
```

```python
genV38_AllImprinted = pd.merge(gencodeV38, imprinted[['symbol', 'allele', 'status']], left_on='symbol', right_on='symbol')
genV38_AllImprinted['score'] = 0
dupSymbol = genV38_AllImprinted['symbol'].duplicated(keep=False)
dupID = genV38_AllImprinted['ID'].duplicated(keep=False)
genV38_AllImprinted = genV38_AllImprinted.loc[~(dupSymbol | dupID)]
```

```python
genV38_AllImprinted['rgb'] = genV38_AllImprinted['allele'].apply(defineAllele)
```

```python
# Removed
allSymbols = imprinted.loc[imprinted['status'] == 'Imprinted', 'symbol']
mappedSymbols = genV38_AllImprinted.loc[genV38_AllImprinted['status'] == 'Imprinted', 'symbol']
print(set(allSymbols) - set(mappedSymbols))
```

```python
# Get only those with true Imprinted status
genV38_NotImprinted = genV38_AllImprinted.loc[genV38_AllImprinted['status'] != 'Imprinted']
genV38_Imprinted = genV38_AllImprinted.loc[genV38_AllImprinted['status'] == 'Imprinted']
```

```python
nRemoved = len(imprinted.loc[imprinted['status'] == 'Imprinted']) - len(genV38_Imprinted)
print('Total', len(imprinted.loc[imprinted['status'] == 'Imprinted']), 'Removed:', nRemoved, 'Retained:', len(genV38_Imprinted))
```

```python
columns = ['chrom', 'start', 'end', 'symbol', 'score', 'orientation', 'start', 'end', 'rgb']
BedTool.from_dataframe(genV38_Imprinted[columns]).sort().moveto('Imprinted-GeneImprint-GRCh37.bed')
BedTool.from_dataframe(genV38_NotImprinted[columns]).sort().moveto('NotImprinted-GeneImprint-GRCh37.bed')
```

```python
plt.rcParams['figure.figsize'] = (16, 9)
plt.rcParams['font.size'] = 20

GM12878set = set(genV38_GM12878.loc[genV38_GM12878['ASE'] != 'Unassigned', 'ID'])
IMR90set = set(genV38_IMR90['ID'])
Imprintedset = set(genV38_Imprinted['ID'])
H1set = set(genV38_H1['ID'])

labels = venn.get_labels([GM12878set, IMR90set, H1set, Imprintedset], fill=['number'])
fig, ax = venn.venn4(labels, names=['GM12878', 'IMR90', 'H1hESC', 'Imprinted'])
fig.tight_layout()
name = 'ASEgenes-overlap-venn'
fig.savefig(f'{name}.svg', dpi=300, bbox_inches='tight')
```

```python
# Here we label each imprinted gene and whether it overlaps an ASTAD in each cell
imprinted = BedTool('Imprinted-GeneImprint-GRCh37.bed').to_dataframe()[['chrom', 'start', 'end', 'name']]
cells = ['GM12878', 'IMR90', 'H1hESC']
for cell in cells:
    ASTAD = f'../diffTAD/results/{cell}/alleleGRCh37/{cell}_a1-vs-{cell}_a2-all-20000-SNPsplit_diff_tad-ASTAD-autosome.bed'
    overlaps = BedTool('Imprinted-GeneImprint-GRCh37.bed').intersect(ASTAD, wa=True, u=True).to_dataframe()['name']
    imprinted[cell] = imprinted['name'].isin(overlaps)
imprinted['OverlapCount'] = imprinted[cells].sum(axis=1)
imprinted.to_csv('Imprinted-overlapASTAD.csv', index=False)
```
