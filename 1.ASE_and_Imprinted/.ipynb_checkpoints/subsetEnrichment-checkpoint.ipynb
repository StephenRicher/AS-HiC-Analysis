Count number of ASE / Imprinted genes overlapping ASTADs and compare to random sample

```python
import numpy as np
import pandas as pd
import seaborn as sns
from pybedtools import BedTool
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from statsmodels.stats.multitest import multipletests
```

```python
plt.rcParams['font.size'] = '15'
```

```python
def getSymbol(x):
    "Extract gene name from GFF3 attributes string"
    attribute = dict([pair.split('=') for pair in x.split(';')])
    return attribute['gene_name']
```

```python
def getStatus(x):
    if x == '255,0,0':
        return 'paternal'
    elif x == '0,0,255':
        return 'maternal'
    elif x == '128,128,128':
        return 'biallelic'
    else:
        return 'monoallelic'
```

```python
index = ['seqname', 'start', 'end']
drop = ['source', 'feature', 'score', 'strand', 'frame', 'attributes']
gencodeGenes = BedTool('../gencode.v38lift37-genes.gff3')
```

#### Process GM12878

```python
gmASTAD = '../diffTAD/results/GM12878/alleleGRCh37/GM12878_a1-vs-GM12878_a2-all-20000-SNPsplit_diff_tad-ASTAD-autosome.bed'

genesOverlapASTAD = gencodeGenes.intersect(gmASTAD, wa=True).to_dataframe().drop_duplicates()
genesOverlapASTAD['GM12878_ASTAD'] = True
genesNotOverlapASTAD = gencodeGenes.intersect(gmASTAD, wa=True, v=True).to_dataframe().drop_duplicates()
genesNotOverlapASTAD['GM12878_ASTAD'] = False

allGenesGM_ASTAD = pd.concat([genesOverlapASTAD, genesNotOverlapASTAD])

allGenesGM_ASTAD['symbol'] = allGenesGM_ASTAD['attributes'].apply(getSymbol)
allGenesGM_ASTAD = allGenesGM_ASTAD.drop(drop, axis=1)
allGenesGM_ASTAD['seqname'] = allGenesGM_ASTAD['seqname'].astype(str)
allGenesGM_ASTAD = allGenesGM_ASTAD.set_index(index)
```

```python
gmTAD = '../diffTAD/results/GM12878/alleleGRCh37/GM12878_a1-vs-GM12878_a2-all-20000-SNPsplit_diff_tad-allTAD-autosome.bed'

genesOverlapTAD = gencodeGenes.intersect(gmTAD, wa=True).to_dataframe().drop_duplicates()
genesOverlapTAD['GM12878_TAD'] = True
genesNotOverlapTAD = gencodeGenes.intersect(gmTAD, wa=True, v=True).to_dataframe().drop_duplicates()
genesNotOverlapTAD['GM12878_TAD'] = False

allGenesGM_TAD = pd.concat([genesOverlapTAD, genesNotOverlapTAD])

allGenesGM_TAD = allGenesGM_TAD.drop(drop, axis=1)

allGenesGM_TAD['seqname'] = allGenesGM_TAD['seqname'].astype(str)
allGenesGM_TAD = allGenesGM_TAD.set_index(index)
```

```python
allGenesGM = pd.merge(allGenesGM_TAD, allGenesGM_ASTAD, left_index=True, right_index=True).drop_duplicates()
```

#### Process IMR90

```python
imASTAD = '../diffTAD/results/IMR90/alleleGRCh37/IMR90_a1-vs-IMR90_a2-all-20000-SNPsplit_diff_tad-ASTAD-autosome.bed'

genesOverlapASTAD = gencodeGenes.intersect(imASTAD, wa=True).to_dataframe().drop_duplicates()
genesOverlapASTAD['IMR90_ASTAD'] = True
genesNotOverlapASTAD = gencodeGenes.intersect(imASTAD, wa=True, v=True).to_dataframe().drop_duplicates()
genesNotOverlapASTAD['IMR90_ASTAD'] = False

allGenesIM_ASTAD = pd.concat([genesOverlapASTAD, genesNotOverlapASTAD]).drop(drop, axis=1)

allGenesIM_ASTAD['seqname'] = allGenesIM_ASTAD['seqname'].astype(str)
allGenesIM_ASTAD = allGenesIM_ASTAD.set_index(index)
```

```python
imTAD = '../diffTAD/results/IMR90/alleleGRCh37/IMR90_a1-vs-IMR90_a2-all-20000-SNPsplit_diff_tad-allTAD-autosome.bed'

genesOverlapTAD = gencodeGenes.intersect(imTAD, wa=True).to_dataframe().drop_duplicates()
genesOverlapTAD['IMR90_TAD'] = True
genesNotOverlapTAD = gencodeGenes.intersect(imTAD, wa=True, v=True).to_dataframe().drop_duplicates()
genesNotOverlapTAD['IMR90_TAD'] = False

allGenesIM_TAD = pd.concat([genesOverlapTAD, genesNotOverlapTAD]).drop(drop, axis=1)

allGenesIM_TAD['seqname'] = allGenesIM_TAD['seqname'].astype(str)
allGenesIM_TAD = allGenesIM_TAD.set_index(index)
```

```python
allGenesIM = pd.merge(allGenesIM_TAD, allGenesIM_ASTAD, left_index=True, right_index=True)
```

#### Combine cell line data

```python
allGenes = pd.merge(allGenesGM, allGenesIM, left_index=True, right_index=True).reset_index()
allGenes['seqname'] = allGenes['seqname'].astype(str)
allGenes = allGenes.set_index(index)
```

```python
gmASEG = 'GM12878-all-31740818-GRCh37.bed'

names = {'chrom': str, 'start': int, 'end': int, 'symbol': 'str', 'status': 'str'}
gm = pd.read_csv(gmASEG, usecols=[0, 1, 2, 3, 8], names=names.keys(), dtype=names, sep='\t')
gm['status'] = gm['status'].apply(getStatus)
gmASEG = gm.loc[gm['status'].isin(['paternal', 'maternal', 'monoallelic']), 'symbol']

allGenes['GM12878_ASEG'] = allGenes['symbol'].isin(gmASEG)
```

```python
imASEG = 'IMR90-ASEG-GSM438363_GSM432687-GRCh37.bed'

names = {'chrom': str, 'start': int, 'end': int, 'symbol': 'str', 'status': 'str'}
im = pd.read_csv(imASEG, usecols=[0, 1, 2, 3, 8], names=names.keys(), dtype=names, sep='\t')
im['status'] = im['status'].apply(getStatus)
imASEG = im.loc[im['status'].isin(['paternal', 'maternal', 'monoallelic']), 'symbol']

allGenes['IMR90_ASEG'] = allGenes['symbol'].isin(imASEG)
```

```python
imprinted = 'Imprinted-GeneImprint-GRCh37.bed'

names = {'chrom': str, 'start': int, 'end': int, 'symbol': 'str'}
imprinted = pd.read_csv(imprinted, usecols=[0, 1, 2, 3], names=names.keys(), dtype=names, sep='\t')

allGenes['Imprinted'] = allGenes['symbol'].isin(imprinted['symbol'])
```

```python
results = {}
rawP = [] # Store in list for post correction
for cell in ['GM12878', 'IMR90']:
    # Consider only genes within a TAD
    genesInTad = allGenes.loc[allGenes[f'{cell}_TAD']]
    for geneType in [f'{cell}_ASEG', 'Imprinted']:
        obs = pd.crosstab(genesInTad[f'{cell}_ASTAD'], genesInTad[geneType])
        chi2, p, dof, exp = chi2_contingency(obs)
        name = 'ASEG' if geneType != 'Imprinted' else geneType
        results[(cell, name)] =  [chi2, p, np.log2(obs / exp)]
        rawP.append(p)
_, adjP, _, _ = multipletests(rawP, method='bonferroni')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 14), sharex=True, sharey=True)
axes = axes.flatten()
for i, (cell, name) in enumerate(results.keys()):
    chi2, p, obsExp = results[(cell, name)]
    p = adjP[i] # Obtain corrected p
    sns.heatmap(
        obsExp, cmap='bwr', center=0, vmin=-1, vmax=1, annot=True, 
        cbar_kws={'label': 'log2(obs / exp)'}, ax=axes[i])
    if i % 2 == 0:
        axes[i].set_ylabel(f'ASTAD')
    else:
        axes[i].set_ylabel('')
    if i > 1:
        axes[i].set_xlabel(f'{name}')
    else:
        axes[i].set_xlabel('')
    axes[i].set_title(f'{cell} - {name}, p(adj) = {p:.3f}', loc='left')
fig.tight_layout()
fig.savefig('ASGeneASTADchi2.svg')
```
