```python
import os
import glob
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from statsmodels.stats.multitest import fdrcorrection
```

```python
plt.rcParams['font.size'] = 20
orderBy = 'maxRnk'
value = 'maxRnk'
```

```python
LOLApaths = glob.glob('results/*fullVCF-LOLA-results.csv')
LOLAresultsAll = pd.concat([pd.read_csv(file) for file in LOLApaths])

LOLAresultsAll['fdr'] = fdrcorrection(LOLAresultsAll['p'])[1]

# Define order by chosen rank across cell types
orderPivot = pd.pivot_table(LOLAresultsAll, index='description', columns='cellType', values=orderBy)
order = orderPivot.max(axis=1).sort_values().index

LOLAresults = (pd.pivot_table(
    LOLAresultsAll, index='description', columns='cellType', values=value)
    .reindex(order))
```

```python
fig, ax = plt.subplots(figsize=(16,9))
sns.heatmap(
    data=LOLAresults, vmin=0, yticklabels=1, annot=True, fmt=".0f", linewidths=.5,
    cmap='Reds_r', cbar_kws={'label': 'Max LOLA Rank'}, ax=ax)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_facecolor('grey')
for index, row in LOLAresultsAll[LOLAresultsAll['fdr'] < 0.05].iterrows():
    namePos = LOLAresults.index.get_loc(row['description'])
    idPos = LOLAresults.columns.get_loc(row['cellType'])
    ax.add_patch(Rectangle((idPos, namePos), 1, 1, ec='Black', alpha=0.5, fc='none', hatch='/', lw=3))
fig.tight_layout()
fig.savefig('LOLAenrichment.svg', dpi=300)
```
