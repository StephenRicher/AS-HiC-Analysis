```python
import os
import glob
import numpy as np
import pandas as pd
```

```python
orderBy = 'maxRnk'
value = 'pValueLog'

LOLAresults = pd.concat([pd.read_csv(file) for file in glob.glob('results/*-LOLA-results.csv')])

# Define order by mean rank across cell types
order = (
    pd.pivot_table(
        LOLAresults, index='description', columns='cellType', values='meanRnk')
    .mean(axis=1)
    .sort_values()
    .index
)

LOLAresults = (pd.pivot_table(
    LOLAresults, index='description', columns='cellType', values=value)
               .reindex(order))
```

```python
pd.set_option('precision', 2)
```

```python
threshold = -np.log10(0.05)
(LOLAresults.style.background_gradient(cmap='Reds', axis=None)
 .applymap(lambda x: 'background-color:white' if x < threshold else ''))
```
