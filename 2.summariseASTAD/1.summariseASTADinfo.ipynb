```python
import sys
sys.path.append("..")
from utilities import readDiffTAD, formatCell, defaultPlotting
import os
import sys
import gzip
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
frequencyData = []
allTADs = []
for cell in ['GM12878', 'IMR90', 'H1hESC']:
    TADs = readDiffTAD(cell, diffOnly=False, X=True, pyBed=False)
    TADbyChrom = TADs.groupby('chrom').size().reset_index().rename({0: 'TADcount'}, axis=1)
    TADbyChrom['TADprop'] = TADbyChrom['TADcount'] / TADbyChrom['TADcount'].sum()
    
    ASTADs = TADs.loc[TADs['type'] == 'ASTAD']
    ASTADbyChrom = ASTADs.groupby('chrom').size().reset_index().rename({0: 'ASTADcount'}, axis=1)
    
    TADbyChrom = pd.merge(ASTADbyChrom, TADbyChrom)
    TADbyChrom['expASTADcount'] = TADbyChrom['TADprop'] *  TADbyChrom['ASTADcount'].sum()
    
    TADbyChrom['ASTADfreq'] = TADbyChrom['ASTADcount'] / TADbyChrom['expASTADcount']
    

    TADbyChrom = TADbyChrom.set_index(['chrom']).dropna().reset_index()
    TADbyChrom['cell'] = cell
    frequencyData.append(TADbyChrom)
    
    TADs = TADs.groupby(['chrom', 'type']).size().reset_index().rename({0: 'count'}, axis=1)
    TADs['cell'] = cell
    allTADs.append(TADs)

allTADs = pd.concat(allTADs)
order = ([
    'chr1', 'chr2', 'chr3', 'chr4', 'chr5', 'chr6', 'chr7', 'chr8', 'chr9', 'chr10', 
    'chr11', 'chr12', 'chr13', 'chr14', 'chr15', 'chr16', 'chr17', 'chr18', 
    'chr19', 'chr20', 'chr21', 'chr22', 'chrX'
])
frequencyData = pd.concat(frequencyData)
frequencyData['ASTADlog2'] = np.log2(frequencyData['ASTADfreq'])
frequencyHeatmap = frequencyData.pivot(index='chrom', columns='cell', values='ASTADlog2').reindex(order)
frequencyHeatmap.columns = [formatCell(c) for c in frequencyHeatmap.columns]
```

# Process HiC info data

```python
hicinfo = defaultdict(list)
with open('allHiC-contactSum.txt') as fh:
    for line in fh:
        line = line.strip()
        if (not line) or line.startswith('#') or line.startswith('Chromosomes'):
            continue
        name, value = line.strip().split('\t')
        if name == 'File:':
            basename = os.path.basename(value).split('-')
            hicinfo['cell'].append(basename[0].split('_')[0])
            if basename[0].endswith('_a1'):
                hicinfo['phaseMode'].append('A1')
            elif basename[0].endswith('_a2'):
                hicinfo['phaseMode'].append('A2')
            else:
                hicinfo['phaseMode'].append('Full')
            hicinfo['chrom'].append(basename[1].split('.')[0])
        elif name == 'Sum of matrix:':
            hicinfo['contacts'].append(int(float(value)))
        elif name == 'Bin_length:':
            hicinfo['binLength'].append(int(float(value)))
        elif name == 'Size:':
            value = ''.join([x for x in value if x != ','])
            hicinfo['Size'].append(int(float(value)))
hicinfo = pd.DataFrame(hicinfo)
hicinfo['totalLength'] = hicinfo['Size'] * hicinfo['binLength']
hicinfo['density'] = hicinfo['contacts'] / (hicinfo['totalLength'] / 10000)
```

```python
ASratio = pd.read_pickle('../1.processASratio/AS-ratio.pkl')
```

```python
hicinfo = pd.merge(hicinfo, allTADs, left_on=['cell', 'chrom'], right_on=['cell', 'chrom'])
hicinfo = pd.merge(hicinfo, ASratio, left_on=['cell', 'chrom'], right_on=['cell', 'chrom'])
hicinfo['phaseMode'] = hicinfo['phaseMode'].apply(lambda x: x if x == 'Full' else 'Allele')
hicinfo = hicinfo.groupby(['cell', 'chrom', 'type', 'phaseMode']).mean().reset_index()
hicinfo = hicinfo.sort_values(['type', 'phaseMode'], ascending=False)
```

```python
names = {'chrom': str, 'size': int}
chromSizes = pd.read_csv('../annotation/hg19.chrom.sizes', names=names.keys(), dtype=names, sep='\t')
chromSizes['mb'] = chromSizes['size'] / 1000000
```

```python
hicinfo = pd.merge(hicinfo, chromSizes, left_on='chrom', right_on='chrom')
```

```python
hicinfo['countPerMb'] = hicinfo['count'] / hicinfo['mb']
hicinfo = hicinfo.loc[hicinfo['type'] == 'ASTAD']
```

```python
hicinfo['cell'] = hicinfo['cell'].apply(lambda x: formatCell(x))
```
