```python
import sys
sys.path.append("..")
from utilities import readDiffTAD, formatCell, defaultPlotting, sampleByGroup
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import matplotlib.lines as lines
from collections import defaultdict
from scipy.ndimage import interpolation
```

```python
width = 4
resolution = 10000 # Resolution of Bedgraph
domainSize = 10 # Number of pixels per TAD
size = domainSize * ((width * 2) + 1 + 2) # Add 2 lengths of either side for edge effects
```

```python
boundary = [None] * (width * domainSize - 1)
internal = ['|'] + ([None] * (domainSize - 2)) + ['|']
ticks = [str(-width)] + boundary + internal + boundary + [str(-width)]
tickIdx = indices = [i for i, x in enumerate(ticks) if x == '|']
```

```python
allData = []
cells = ['GM12878', 'H1hESC', 'IMR90']
for cell in cells:
    data = readDiffTAD(cell, diffOnly=False, X=False, pyBed=False)
    data['cell'] = cell
    allData.append(data)
allData = pd.concat(allData)
allData['size'] = allData['end'] - allData['start']
allData['roundSize'] = ((allData['size'] // 100000) * 100000) + 100000

allData = allData.reset_index(drop=True)
```

```python
defaultPlotting(ratio=1/3)
fig, axes = plt.subplots(1, 3, sharey=True)
axes = axes.flatten()
for i, cell in enumerate(cells):
    ax = axes[i]
    sns.histplot(data=allData.loc[allData['cell']==cell], 
                 x='size', hue='type', stat='probability', 
                 common_norm=False, log_scale=True, ax=ax)
    ax.set_xlabel('Domain Size (bp)')
    ax.set_title(formatCell(cell), loc='left')
    handles, labels = ax.get_legend_handles_labels()
    ax.get_legend().set_title(None)
    if i < 2:
        ax.get_legend().remove()
    
fig.tight_layout()
axes[0].text(
    0, 1, 'abcdefghijk', transform=fig.transFigure, 
    fontsize='large', verticalalignment='top', weight='bold')
fig.savefig('DomainSizeDistribution.svg', transparent=True)
```

Constrain domain sizes limits to equal ranges for ASTADs and TADs.
Domain limits set to the inner 5% - 95% quantile range of ASTADs / TADs

```python
allTrackData = {}
tracks = ([
    'Insulation', 'CTCF', 'Cscore', 'GC', 'Mappability', 'Het_SNPs', 'Het_Indels', 'ASM',
    'WGBS', 'ATAC', 'ASM'
])

tracks = ['Insulation','CTCF', 'Cscore', 'GC', 'Mappability', 'Het_SNPs', 'Het_Indels', 'ASM']
for cell in ['GM12878', 'IMR90', 'H1hESC']:

    sampleGroups = defaultdict(int, df.loc[df['type']=='ASTAD'].groupby(['chrom', 'roundSize']).size().to_dict())
    df = allData.loc[allData['cell'] == cell]
    lowerLimit = df.groupby('type')['size'].quantile(0.05).max()
    upperLimit = df.groupby('type')['size'].quantile(0.95).min()
    df = df[df['size'].between(lowerLimit, upperLimit)]

    for track in tracks:
        
        print(cell, track)
        # Read bedgraph
        names = {'chrom': str, 'start': int, 'end': int, 'score': float}
        if track in ['GC', 'Mappability']:
            file = f'{track}-10kb-hg19.bedgraph'
        else:
            file = f'{cell}-{track}-10kb-hg19.bedgraph'
        bedgraph = pd.read_csv(file, names=names.keys(), dtype=names, sep='\t')
        bedgraph['chrom'] = bedgraph['chrom'].apply(lambda x: f'chr{x}' if not x.startswith('chr') else x)
        bedgraph = bedgraph.set_index(['chrom', 'start'])
        # Process data
        indexMap = {}
        ASTADindexes = []
        for row in df.itertuples():
            startBin = resolution * (row.start // resolution)
            endBin = resolution * (row.end // resolution)

            TADbins = [(str(row.chrom), pos) for pos in range(startBin, endBin + 1, resolution)]
            # +1 to add buffer for interpolation
            nBins = len(TADbins) * resolution * (width + 1)
            # Retrieve equal number of bins before and after TAD
            beforeTAD = [(str(row.chrom), pos) for pos in range(startBin - nBins, startBin, resolution)]
            afterTAD = [(str(row.chrom), pos) for pos in range(endBin + resolution, endBin + nBins + 1, resolution)]
            allBins =  beforeTAD + TADbins + afterTAD

            try:
                scores = np.array(bedgraph.loc[allBins, 'score'])
            except KeyError: # Out of chromosome bounds
                continue
            # Record valid indexes of ASTADs
            if row.type == 'ASTAD':
                ASTADindexes.append(row.Index)
            z = size / len(scores)
            # Store interpolation by index
            rescaled = interpolation.zoom(scores, z)
            # Trim interpolation by 1 domainSize to remove edge effects
            indexMap[row.Index] = rescaled[domainSize: -domainSize]
        indexMap = pd.DataFrame(indexMap).T

        # Compute Z scrores
        obsRef = indexMap.loc[ASTADindexes].mean(axis=0)
        nASTADs = len(ASTADindexes)
        samples = []
        for i in range(1000):
            sampledIndex = sampleByGroup(
                df.loc[indexMap.index], ['chrom', 'roundSize'], sampleGroups, replace=True)
            sampledIndex = list(sampledIndex.index.get_level_values(2))
            samples.append(indexMap.loc[sampledIndex].mean(axis=0).values)
            #samples.append(indexMap.sample(nASTADs).mean(axis=0).values)
        sampleMean = np.array(samples).mean(axis=0)
        sampleStd = np.array(samples).std(axis=0)
        zRef = (obsRef - sampleMean) / sampleStd
        # Store zRef in dataframe for heatmap
        zRef = pd.DataFrame({'pos': range(len(zRef)), '': zRef}).set_index('pos').T

        # Compute raw mean proportions
        TADscores = indexMap.T.unstack().reset_index(level=1).rename({'level_1': 'pos', 0: 'score'}, axis=1)
        scoredData = pd.merge(df, TADscores, left_index=True, right_index=True)
        proportion = scoredData.groupby(['type', 'pos'])['score'].mean().reset_index()
    
        zRef.to_pickle(f'data/{cell}-{track}-zRef.pkl')
        proportion.to_pickle(f'data/{cell}-{track}-proportion.pkl')
        allTrackData[(cell, track)] = [zRef, proportion]
```

```python
plt.rcParams['font.size'] = 16
```

```python
for i, ((cell, track), (zRef, proportion)) in enumerate(allTrackData.items()):
    fig, axes, = plt.subplots(1, 2, figsize=(16, 9))
    sns.heatmap(zRef, xticklabels=ticks, center=0, cmap='bwr', ax=axes[0])
    sns.lineplot(x='pos', y='score', hue='type', data=proportion, ax=axes[1])
    for n, ax in enumerate(axes):
        ax.set_xlabel('Scaled position along domain')
        ax.tick_params(left=False, bottom=False)
        if n % 2 == 0:
            ax.set_ylabel("ASTAD vs. non-ASTAD (Z-score)")
            ax.set_title(f'{formatCell(cell)}, {track}', loc='left')
        else:
            ax.axvline(tickIdx[0], color='black', linestyle='--', alpha=0.5)
            ax.axvline(tickIdx[1], color='black', linestyle='--', alpha=0.5)
            ax.set_xticks(range(len(ticks)))
            ax.set_xticklabels(ticks)
            ax.set_ylabel(f'{track} Signal')
            # Remove legend title
            handles, labels = ax.get_legend_handles_labels()
            ax.get_legend().set_title(None)
    fig.tight_layout()
    fig.savefig(f'plots/{cell}-{track}-ASTADexploration.svg', dpi=600)
```
