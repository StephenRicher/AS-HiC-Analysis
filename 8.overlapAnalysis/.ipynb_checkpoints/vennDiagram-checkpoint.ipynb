```python
import os
import sys
sys.path.append("..")
import numpy as np
import pandas as pd
import seaborn as sns
from pybedtools import BedTool
import matplotlib.pyplot as plt
from collections import defaultdict
from matplotlib_venn import venn3
from utilities import readDiffTAD
```

```python
plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (20, 20)
```

```python
def intersectBoth(a, b, c, fraction=0.9):
    """ Return Bed intervals in both pairs """
    a_and_b = a.intersect(b, f=fraction, r=True, wb=True)
    a_and_c = a.intersect(c, f=fraction, r=True, wb=True)
    b_and_c = b.intersect(c, f=fraction, r=True, wb=True)
    
    return a_and_b, a_and_c, b_and_c


def computeIntersectCount(a, b, c, fraction=0.9):
    a_and_b, a_and_c, b_and_c = intersectBoth(a, b, c, fraction)
    # Get centre count
    a_and_b_and_c = a_and_b.intersect(c, f=fraction, r=True, wb=True).count()

    a_and_b_not_c = a_and_b.count() - a_and_b_and_c
    a_and_c_not_b = a_and_c.count() - a_and_b_and_c
    b_and_c_not_a = b_and_c.count() - a_and_b_and_c
    a_not_b_not_c = a.count() - a_and_b_not_c - a_and_c_not_b - a_and_b_and_c
    b_not_c_not_a = b.count() - a_and_b_not_c - b_and_c_not_a - a_and_b_and_c
    c_not_b_not_a = c.count() - a_and_c_not_b - b_and_c_not_a - a_and_b_and_c

    return (a_not_b_not_c, b_not_c_not_a, a_and_b_not_c, 
            c_not_b_not_a, a_and_c_not_b, b_and_c_not_a, 
            a_and_b_and_c)
```

```python
fraction = 0.90
```

```python
a = readDiffTAD('GM12878', X=False, pyBed=True).to_dataframe()
a['start'] = a['start'] - 5000
a['end'] = a['end'] - 5000
a = BedTool.from_dataframe(a)
b = readDiffTAD('IMR90', X=False, pyBed=True)
c = readDiffTAD('H1hESC', X=False, pyBed=True)
```

```python
a_not_b_not_c, b_not_c_not_a, a_and_b_not_c, c_not_b_not_a, a_and_c_not_b, b_and_c_not_a, a_and_b_and_c = (
    computeIntersectCount(a, b, c))
```

```python
aAS = readDiffTAD('GM12878', diffOnly=True, X=False, pyBed=True).to_dataframe()
aAS['start'] = aAS['start'] - 5000
aAS['end'] = aAS['end'] - 5000
aAS = BedTool.from_dataframe(aAS)
bAS = readDiffTAD('IMR90', diffOnly=True, X=False, pyBed=True)
cAS = readDiffTAD('H1hESC', diffOnly=True, X=False, pyBed=True)
```

```python
a_not_b_not_cAS, b_not_c_not_aAS, a_and_b_not_cAS, c_not_b_not_aAS, a_and_c_not_bAS, b_and_c_not_aAS, a_and_b_and_cAS = (
    computeIntersectCount(aAS, bAS, cAS))
```

```python
subsets = (
    a_not_b_not_c, b_not_c_not_a, a_and_b_not_c, c_not_b_not_a, 
    a_and_c_not_b, b_and_c_not_a, a_and_b_and_c)
total = sum(subsets)

subsetsAS = (
    a_not_b_not_cAS, b_not_c_not_aAS, a_and_b_not_cAS, c_not_b_not_aAS, 
    a_and_c_not_bAS, b_and_c_not_aAS, a_and_b_and_cAS)
totalAS = sum(subsetsAS)

fig, axes = plt.subplots(2, 2)
axes = axes.flatten()
venn3(subsets=subsets, set_labels=('GM12878', 'IMR90', 'H1hESC'), ax=axes[0])
venn3(subsets=subsets, set_labels=('GM12878', 'IMR90', 'H1hESC'), 
      subset_label_formatter=lambda x: f'{(x / total):1.1%}', ax=axes[1])
venn3(subsets=subsetsAS, set_labels=('GM12878', 'IMR90', 'H1hESC'), ax=axes[2])
venn3(subsets=subsetsAS, set_labels=('GM12878', 'IMR90', 'H1hESC'), 
      subset_label_formatter=lambda x: f'{(x / totalAS):1.1%}', ax=axes[3])
fig.tight_layout()
fig.savefig(f'TAD-ASTAD-overlap-venn.svg', dpi=300)
```

```python
nA = aAS.count()
nB = bAS.count()
nC = cAS.count()

# Get observed count
a_and_bObs, a_and_cObs, b_and_cObs = intersectBoth(aAS, bAS, cAS, fraction)
a_and_b_and_cObs = a_and_bObs.intersect(cAS, f=fraction, r=True, wb=True).count()
ABobs = a_and_bObs.count()
ACobs = a_and_cObs.count()
BCobs = b_and_cObs.count()
ABCobs = a_and_b_and_cObs

ABexp = []
ACexp = []
BCexp = []
ABCexp = []
for i in range(10000):
    Asample = BedTool.from_dataframe(a.to_dataframe().sample(nA)).sort()
    Bsample = BedTool.from_dataframe(b.to_dataframe().sample(nB)).sort()
    Csample = BedTool.from_dataframe(c.to_dataframe().sample(nC)).sort()
    
    a_and_bSample, a_and_cSample, b_and_cSample = intersectBoth(Asample, Bsample, Csample, fraction)
    a_and_b_and_cSample = a_and_bSample.intersect(Csample, f=fraction, r=True, wb=True).count()
    ABexp.append(a_and_bSample.count())
    ACexp.append(a_and_cSample.count())
    BCexp.append(b_and_cSample.count())
    ABCexp.append(a_and_b_and_cSample)
ABexp = np.array(ABexp)
ACexp = np.array(ACexp)
BCexp = np.array(BCexp)
ABCexp = np.array(ABCexp)

ABz = (ABobs - ABexp.mean()) / ABexp.std()
ACz = (ACobs - ACexp.mean()) / ACexp.std()
BCz = (BCobs - BCexp.mean()) / BCexp.std()
ABCz = (ABCobs - ABCexp.mean()) / ABCexp.std()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,12))
plt.rcParams['font.size'] = 16
axes = axes.flatten()
# AB: GM12878 - IMR90
sns.histplot(x=ABexp, stat='probability', ax=axes[0])
axes[0].set_xlabel('Expected Overlap')
axes[0].axvline(ABobs, color='red')
axes[0].set_title(f'GM12878 - IMR90 (z = {ABz:.1f})', loc='left')
# AC: GM12878 - H1hESC
sns.histplot(x=ACexp, stat='probability', ax=axes[1])
axes[1].set_xlabel('Expected Overlap')
axes[1].axvline(ACobs, color='red')
axes[1].set_title(f'GM12878 - H1hESC (z = {ACz:.1f})', loc='left')
# BC: IMR90 - H1hESC
sns.histplot(x=BCexp, stat='probability', ax=axes[2])
axes[2].set_xlabel('Expected Overlap')
axes[2].axvline(BCobs, color='red')
axes[2].set_title(f'IMR90 - H1hESC (z = {BCz:.1f})', loc='left')
# ABC: GM12878 - IMR90 - H1hESC
sns.histplot(x=ABCexp, stat='probability', ax=axes[3])
axes[3].set_xlabel('Expected Overlap')
axes[3].axvline(ABCobs, color='red')
axes[3].set_title(f'GM12878 - IMR90 - H1hESC (z = {ABCz:.1f})', loc='left')

fig.tight_layout()
fig.savefig('Observed-vs-Expected-ASTADoverlap.svg')
```

```python
genes = BedTool('../gencode.v38lift37-genes.gff3')
atleast2 = a_and_bObs.cat(a_and_cObs, b_and_cObs).sort().merge()
overlappingGenes = genes.intersect(atleast2, wa=True, u=True)
```

```python
allAttributes = {'ID': [], 'gene_id': [], 'gene_type': [], 'gene_name': []}
for i in overlappingGenes.to_dataframe()['attributes']:
    attributes = i.split(';')
    for attr in attributes:
        attr = attr.split('=')
        name = attr[0]
        value = '='.join(attr[1:])
        if name in allAttributes.keys():
            allAttributes[name].append(value)
allAttributes = pd.DataFrame(allAttributes)
allAttributes['ensembl'] = allAttributes['ID'].apply(lambda x: x.split('.')[0])
allAttributes['ensembl'].to_csv('2way-ASTADoverlap-geneList.txt', index=False, header=False)
```

```python
unmappedIDs = pd.read_csv('unmappedIDs-panther.txt', squeeze=True)
unmapped = allAttributes[allAttributes['ensembl'].isin(unmappedIDs)]
unmapped.groupby('gene_type').count()
```

```python
mapped = allAttributes[~allAttributes['ensembl'].isin(unmappedIDs)]
mapped.groupby('gene_type').count()
```
