```python
import glob
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from statsmodels.stats.multitest import multipletests
```

```python
plt.rcParams['font.size'] = '16'
```

```python
files = glob.glob(f'*-SNP-overlapSummary.txt')
files.reverse()
```

```python
chi = {}

for file in files:
    cell1, cell2 = file.split('-')[:2]
    obs = pd.read_csv(file, sep='\t', index_col=0)
    chi2, p, dof, exp = chi2_contingency(obs)
    ## Temporary rename ##
    if cell1 == 'H1hESC':
        cell1 = 'H1-hESC'
    elif cell1 == 'IMR90':
        cell1 = 'IMR-90'
    if cell2 == 'H1hESC':
        cell2 = 'H1-hESC'
    elif cell2 == 'IMR90':
        cell2 = 'IMR-90'
    chi[(cell1, cell2)] = (p, obs / exp)
rawP = [v[0] for v in chi.values()]
_, adjP, _, _ = multipletests(rawP, method='bonferroni')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
axes = axes.flatten()
for i, ((cell1, cell2), (p, obsExp)) in enumerate(chi.items()):
    obsExp = np.log2(obsExp)
    obsExp = obsExp.rename({'ASTAD': 'Conserved ASTAD Interval', 'non-ASTAD': 'Other'}, axis=1)
    sns.heatmap(
        data=obsExp, annot=True, fmt='.3f',
        vmin=-0.15, vmax=0.15, center=0, cmap='bwr', ax=axes[i])
    if i == 3:
        title = f'3-Way Overlap, p(adj) = {adjP[i]:.3f}'
    else:
        title = f'{cell1} & {cell2}, p(adj) = {adjP[i]:.3f}'
    axes[i].set_title(title, loc='left')
fig.tight_layout()
fig.savefig(f'SNPidentity.svg', dpi=300)
```
