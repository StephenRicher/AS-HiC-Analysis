```python
import subprocess
import pandas as pd
import seaborn as sns
from pybedtools import BedTool
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
```

```python
plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (16, 9)
```

```python
GM = BedTool('../diffTAD/results/GM12878/alleleGRCh37/GM12878_a1-vs-GM12878_a2-all-20000-SNPsplit_diff_tad-ASTAD-autosome.bed')
GM = GM.to_dataframe()
GM['start'] = GM['start'] - 5000
GM['end'] = GM['end'] - 5000
GM = BedTool.from_dataframe(GM)
IM = BedTool('../diffTAD/results/IMR90/alleleGRCh37/IMR90_a1-vs-IMR90_a2-all-20000-SNPsplit_diff_tad-ASTAD-autosome.bed')
```

```python
GM_and_IM = GM.intersect(GM, f=0.9, r=True, wb=True).to_dataframe()
GM_and_IM['start'] = GM_and_IM.apply(lambda x: (x['start'] + x['thickEnd']) / 2, axis=1)
GM_and_IM['end'] = GM_and_IM.apply(lambda x: (x['end'] + x['itemRgb']) / 2, axis=1)
GM_and_IM = GM_and_IM[['chrom', 'start', 'end']]
GM_and_IM['group'] = 'Overlap'

GM_not_IM = GM.intersect(IM, f=0.9, v=True, r=True).to_dataframe()[['chrom', 'start', 'end']]
GM_not_IM['group'] = 'GM12878'

IM_not_GM = IM.intersect(GM, f=0.9, v=True, r=True).to_dataframe()[['chrom', 'start', 'end']]
IM_not_GM['group'] = 'IMR90'
```

```python
allData = pd.concat([GM_and_IM, GM_not_IM, IM_not_GM]).drop_duplicates()

GMvcf = '../VCFs/GM12878-hetOnly.vcf.gz'
IMvcf = '../VCFs/IMR90-hetOnly.vcf.gz'

intersect = {'total': [], 'shared': [], 'identical': []}
for i, chrom, start, end, group in allData.itertuples():
    regions = f'{chrom}:{start-10000}-{start+10000},{chrom}:{end-10000}-{end+10000}'
    #regions = f'{chrom}:{start}-{end}'
    subprocess.call([
        'bcftools', 'isec', '-p', 'tmp', '-c', 'all', 
        GMvcf, IMvcf, '--regions', regions
    ])
    total = 0
    shared = 0
    identical = 0
    for i in range(4):
        with open(f'tmp/000{i}.vcf') as fh:
            for line in fh:
                if not line.startswith('#'):
                    if (i < 3):
                        total += 1
                    else:
                        shared += 1
    subprocess.call([
        'bcftools', 'isec', '-p', 'tmp', '-c', 'none', 
        GMvcf, IMvcf, '--regions', regions
    ])
    with open(f'tmp/0003.vcf') as fh:
        for line in fh:
            if not line.startswith('#'):
                identical += 1
    intersect['total'].append(total)
    intersect['shared'].append(shared)
    intersect['identical'].append(identical)
```

```python
allData['total'] = intersect['total']
allData['shared'] = intersect['shared']
allData['identical'] = intersect['identical']
allData['propShared'] = allData['shared'] / allData['total']
allData['propIdentical'] = allData['identical'] / allData['total']
allData['propIdenticalofShared'] = allData['identical'] / allData['shared']
allData['ASTAD overlap'] = allData['group'].apply(lambda x: True if x == 'Overlap' else False)
```

```python
for mode in ['Shared', 'Identical']:
    u, p = mannwhitneyu(
        allData.loc[allData['ASTAD overlap'], f'prop{mode}'], 
        allData.loc[~allData['ASTAD overlap'], f'prop{mode}'])

    fig, ax = plt.subplots()
    sns.histplot(
        data=allData.reset_index(), x=f'prop{mode}', hue='ASTAD overlap', stat='probability', 
        common_norm=False, ax=ax)
    ax.set_xlabel(f'Proportion of {mode.lower()} SNPs between GM12878 and IMR90 within 10kb of TAD boundary')
    ax.set_title(f'p = {p:.3f}', loc='left')
    fig.tight_layout()
    fig.savefig(f'SNPidentity-prop{mode}-GM12878-IMR90.svg', dpi=300)
```
