```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.size'] = '16'
```

```python
names = ({
    'chrom': str, 'start': int, 'end': int, 'class': str, 'TargetGene': str, 
    'ASEG': bool, 'diffDomain': bool, 'GWAS': int, 'cell': str
})
    
data = pd.read_csv(
    'allData-ABC-GWAS-overlapASTAD.bed.gz', 
    skiprows=1, names=names.keys(), dtype=names, sep='\t')
```

```python
gwas = pd.read_csv('gwas_catalog_v1.0-associations_e104_r2021-11-22.tsv.gz', usecols=[7], sep='\t')
gwas['ID'] = gwas.index + 1
```

```python
data = data.loc[data['diffDomain'] & data['ASEG'] & (data['GWAS'] > 0)]
```

```python
data = pd.merge(data, gwas, left_on='GWAS', right_on='ID')
```

```python
data.drop(['ASEG', 'diffDomain', 'ID'], axis=1).to_excel("ASEGtargetGenes.xlsx", index=False) 
```

```python
for cell, df in data.groupby('cell'):
    print(cell)
    genes = list(df['TargetGene'].unique())
    genes = ', '.join(genes)
    print(genes)
```
