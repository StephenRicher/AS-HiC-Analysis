```python
import sys
sys.path.append('..')
from utilities import formatP, formatCell, defaultPlotting
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from statsmodels.stats.multitest import multipletests
```

```python
defaultPlotting(width=180, ratio=0.5)
```

```python
df = pd.read_csv('GWAS-ABC-overlaps.txt', sep='\s+', names=['cell', 'count', 'group'])
df['domain'] = df['group'].apply(lambda x: x.split('-')[0])
df['domain'] = df['domain'].apply(lambda x: 'non-ASTAD' if x == 'notASTAD' else x)
df['overlap'] = df['group'].apply(lambda x: 'Enhancer' if x.split('-')[1] == 'ABC' else 'non-Enhancer')
```

```python
cells = ['GM12878', 'H1hESC', 'IMR90']
```

```python
results = {}
rawP = [] # Store in list for post correction
for cell in cells:
    obs = df.loc[df['cell']==cell].pivot(columns='overlap', index='domain', values='count')
    chi2, p, dof, exp = chi2_contingency(obs, correction=True)
    results[formatCell(cell)] =  [chi2, p, dof, obs, exp]
    rawP.append(p)
_, adjP, _, _ = multipletests(rawP, method='bonferroni')
```

```python
fig, axes = plt.subplots(2,2)
axes = axes.flatten()
cbar_ax = fig.add_axes([.66, .22, .2, .1])
with pd.ExcelWriter(f'allChiData-Enhancer.xlsx') as writer:
    for i, cell in enumerate(results.keys()):
        chi2, p, dof, obs, exp = results[cell]
        pearsonResiduals = (obs - exp) / (np.sqrt(exp))
        p = adjP[i] # Obtain corrected p
        sns.heatmap(
            pearsonResiduals, cmap='bwr', center=0, vmin=-26, vmax=26, annot=True,
            cbar_ax=(None if i else cbar_ax), cbar=(i==0),
            linewidths=1, linecolor='black', 
            cbar_kws={'label': 'Pearson\'s Residuals', 'orientation': 'horizontal'},
            fmt='.2f', ax=axes[i])
        p = formatP(p)
        title = f'{cell}, $x^2$ = {chi2:.2f}, $p_{{adj}}$ {p}'
        axes[i].set_title(title, loc='left')
        axes[i].set_ylabel('')
        axes[i].set_xlabel('')
        axes[i].set_yticklabels(axes[0].get_yticklabels(), fontsize=7)
        axes[i].set_xticklabels(axes[0].get_xticklabels(), fontsize=7)
        obs.to_excel(writer, sheet_name=cell)
axes[-1].axis('off')
fig.tight_layout()
fig.savefig('HetVariantsInABCenhancers.svg')
```

```python
def assignASEG(x):
    if x in list(aseg):
        return 'ASEG'
    else:
        return 'non-ASEG'
```

```python
def collapseStates(x):
    """ ABC enhancer is ASE if it links to atleast 1 ASE.
        ABC is ASTAD if it overlaps an ASTAD anywhere """
    if (x['ASEG'] == 'ASEG').any():
         expression ='ASE'
    else:
         expression ='non-ASE'
    if (x['group'] == 'ASTAD').any():
        tad = 'ASTAD'
    else:
        tad = 'non-ASTAD'
    return tad, expression
```

```python
gencodeGenes = pd.read_pickle('../1.ASE_and_Imprinted/gencode.v38lift37-genes.pkl')
```

```python
gwas = pd.read_csv(
    '../annotation/gwas_catalog_v1.0-associations_e104_r2021-11-22.tsv.gz', 
    usecols=[7], sep='\t').reset_index().rename({'index': 'GWAS'}, axis=1)
```

```python
names = ([
    'chromEnh', 'startEnh', 'endEnd', 'name', 'class', 'activity_base', 
    'TargetGene', 'TargetGeneTSS', 'TargetGeneExpression', 
    'TargetGenePromoterActivityQuantile', 'TargetGeneIsExpressed', 
    'distance', 'isSelfPromoter', 'hic_contact', 'powerlaw_contact', 
    'powerlaw_contact_reference','hic_contact_pl_scaled', 'hic_pseudocount', 
    'hic_contact_pl_scaled_adj', 'ABC.Score.Numerator', 'ABC.Score', 
    'powerlaw.Score.Numerator', 'powerlaw.Score', 'CellType', 'GWAS', 'group'
])
```

```python
dropCols = ([
    'activity_base', 'TargetGeneTSS', 'TargetGeneExpression', 
    'TargetGenePromoterActivityQuantile', 'TargetGeneIsExpressed', 'distance', 
    'isSelfPromoter', 'hic_contact', 'powerlaw_contact', 'powerlaw_contact_reference', 
    'hic_contact_pl_scaled', 'hic_pseudocount', 'hic_contact_pl_scaled_adj', 'ABC.Score', 
    'powerlaw.Score', 'powerlaw.Score.Numerator', 'CellType'
])
```

```python
results = {}
allABC = []
rawP = [] # Store in list for post correction
for cell in cells:
    ABC_in_ASTAD = pd.read_csv(
        f'{cell}/{cell}-ABC_in_ASTAD.bed', names=names, sep='\t').drop(dropCols + ['GWAS'], axis=1)
    ABC_in_ASTAD['group'] = 'ASTAD'
    ABC_Not_in_ASTAD = pd.read_csv(
        f'{cell}/{cell}-ABC_Not_in_ASTAD.bed', names=names, sep='\t').drop(dropCols + ['GWAS'], axis=1)
    ABC_Not_in_ASTAD['group'] = 'nonASTAD'
    ABC = pd.concat([ABC_in_ASTAD, ABC_Not_in_ASTAD]).reset_index(drop=True)

    ABC = pd.merge(gencodeGenes, ABC, left_on='symbol', right_on='TargetGene').drop('TargetGene', axis=1)

    aseg = pd.read_csv(f'../1.ASE_and_Imprinted/{cell}-ASEG-GRCh37.bed', sep='\t', header=None)[3]
    ABC['ASEG'] = ABC['symbol'].apply(assignASEG)

    GWASstats = (pd.read_csv(
            f'{cell}/{cell}-ABC-GWAS-allStats.bed', names=names, sep='\t')
         .drop(dropCols, axis=1)
         .rename({'group': 'GWASgroup'}, axis=1)
    )
    GWASstats = pd.merge(
        gencodeGenes, GWASstats, 
        left_on='symbol', right_on='TargetGene').drop('TargetGene', axis=1)
    
    GWASstats = pd.merge(GWASstats, gwas, left_on='GWAS', right_on='GWAS')

    ABC = pd.merge(
        ABC, GWASstats[['ID', 'name', 'GWAS', 'GWASgroup', 'DISEASE/TRAIT']], 
        left_on=['ID', 'name'], right_on=['ID', 'name'], how='outer')

    ABC['cell'] = cell
    allABC.append(ABC)
    
    # For GM12878 we have confirmed bialleilic, so restrict the test to this list
    if cell == 'GM12878':
        validGenes = pd.read_csv('../1.ASE_and_Imprinted/GM12878-all-GRCh37.bed', sep='\t', header=None)[3]
        subABC = ABC.loc[ABC['symbol'].isin(validGenes)]
    else:
        subABC = ABC.copy()
        
    obs = (
        subABC[['name', 'group', 'ASEG']]
        .drop_duplicates().groupby('name')
        .apply(collapseStates).reset_index()
    )
    obs[['group', 'ASEG']] = obs[0].to_list()
    obs = pd.crosstab(obs['group'], obs['ASEG'])
    chi2, p, dof, exp = chi2_contingency(obs, correction=True)
    results[formatCell(cell)] =  [chi2, p, dof, obs, exp]
    rawP.append(p)
_, adjP, _, _ = multipletests(rawP, method='bonferroni')
allABC = pd.concat(allABC).reset_index(drop=True)
allABC['GWAS'] = allABC['GWAS'].fillna(-1).astype(int)
```

```python
fig, axes = plt.subplots(2,2)
axes = axes.flatten()
cbar_ax = fig.add_axes([.66, .22, .2, .1])
with pd.ExcelWriter(f'allChiData-ASEenhancers.xlsx') as writer:
    for i, cell in enumerate(results.keys()):
        chi2, p, dof, obs, exp = results[cell]
        pearsonResiduals = (obs - exp) / (np.sqrt(exp))
        p = adjP[i] # Obtain corrected p
        sns.heatmap(
            pearsonResiduals, cmap='bwr', center=0, vmin=-4, vmax=4, annot=True,
            cbar_ax=(None if i else cbar_ax), cbar=(i==0),
            linewidths=1, linecolor='black',
            cbar_kws={'label': 'Pearson\'s Residuals', 'orientation': 'horizontal'},
            fmt='.2f', ax=axes[i])
        p = formatP(p)
        title = f'{cell}, $x^2$ = {chi2:.2f}, $p_{{adj}}$ {p}'
        axes[i].set_title(title, loc='left')
        axes[i].set_ylabel('')
        axes[i].set_xlabel('')
        axes[i].set_yticklabels(axes[i].get_yticklabels(), fontsize=7)
        axes[i].set_xticklabels(axes[i].get_xticklabels(), fontsize=7)
        obs.to_excel(writer, sheet_name=cell)
axes[-1].axis('off')
fig.tight_layout()
fig.savefig('ABC-Enhancers-ASEGstatus.svg')
```
